```python
#matplotlib 인코딩
def setting_styles_basic() :
    rcParams['font.family'] = 'malgun Gothic'
    rcParams['axes.unicode_minus'] = False
```

```python
#############구별 주거형태 현황 그래프#####################
#한글 깨짐 방지
setting_styles_basic()

#데이터 전처리
housingtype_dashboard=pd.read_csv('./housingType_dasboard.csv')
housingtype_dashboard.sort_values(['행정구'],inplace=True,ignore_index=True)

#데이터 입력값에 따라 그래프 변화
gu_name=input('구 이름 >>')
mask=housingtype_dashboard['행정구']==gu_name
h_dashboard=housingtype_dashboard[mask]
h_dashboard

x=h_dashboard.values[0][2:].tolist() #주거형태별 갯수
y=h_dashboard.columns[2:].to_list() #주거형태

cmap = plt.get_cmap('Set2')
chart_colors = cmap(np.array([0,1,2,3,4,5,6,7]))

fig = plt.figure(figsize=(10,10)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성 

pie=ax.pie(x,colors=chart_colors,autopct='%.2f')
plt.legend(pie[0],y, loc='upper right',bbox_to_anchor=(1.04,1)) ## 범례 표시
plt.show()
```
