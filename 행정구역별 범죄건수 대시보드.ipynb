```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
def setting_styles_basic() :
    rcParams['font.family'] = 'malgun Gothic'
    rcParams['axes.unicode_minus'] = False
```

```python
setting_styles_basic()
```

```python
crime=pd.read_csv('./data/seoul_crime_cnt.csv')
crime.head()
```

```python
crime=crime.set_index('행정구별',drop=True)
```

```python
crime
```

```python
crime.index[1:][0]
```

```python

x=range(len(crime.index[1:]))
y=crime.iloc[1:,0]
plt.bar(x,y,color='red')

plt.xticks(x,labels=crime.index[1:],rotation=50)

plt.title(crime.index[0])
plt.xlabel('행정구')
plt.ylabel('범죄건수')


plt.show()
```
