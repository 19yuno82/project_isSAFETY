```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
#matplotlib 인코딩
def setting_styles_basic() :
    rcParams['font.family'] = 'malgun Gothic'
    rcParams['axes.unicode_minus'] = False
```

```python
##############행정구별 인구수 그래프######################

#데이터 불러오기
pop_dashboard=pd.read_csv('./data/pop_dashboard.csv')

#데이터 값 따라 대시보드 데이터 변경
gu_name=input('구역 이름 >>')
mask=pop_dashboard['행정구']==gu_name
pop=pop_dashboard[mask]
pop=pop[['총인구수','내국인','외국인']].astype(int)

x=pop.columns.to_list() #x축 데이터
y=pop.iloc[0].to_list() #Y축 데이터

#bar colors
colors=['orange','royalblue','salmon']

plt.bar(x,y,width=0.5,color=colors)
plt.xlabel('인구')
plt.ylabel('수',rotation=0)

#plt.text(x좌표,y좌표,적고자 하는 값)
for i in range(len(x)) :
    plt.text(i-0.1,y[i]-5,f'{y[i]}')
#ylim을 설정해서 막대그래프 확대
#plt.ylim(45,100)

#제목 달아주기
plt.title(gu_name, loc='left')

plt.show()
```
